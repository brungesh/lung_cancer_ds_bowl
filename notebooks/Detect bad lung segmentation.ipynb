```python
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
from skimage import measure
import pandas as pd
wp2 = '/mnt/hd2/preprocessed5/'
sys.path.append('/home/mingot/lung_cancer_ds_bowl/src')

from utils import plotting

file_list = [g for g in os.listdir(wp2)] # if g.startswith('dsb_')]
# file_list = [
#     'dsb_ebd601d40a18634b100c92e7db39f585.npz',
#     'dsb_8c63c8ebd684911de92509a8a703d567.npz',
#     'dsb_380eb569a5750648434cc8ae8da4a0a9.npz',
# ]
file_list = [os.path.join(wp2, f) for f in file_list] # better save the absolute name
print(len(file_list))
print(file_list[0:10])


def bbox2_3D(img):

    r = np.any(img, axis=(1, 2))
    c = np.any(img, axis=(0, 2))
    z = np.any(img, axis=(0, 1))

    rmin, rmax = np.where(r)[0][[0, -1]]
    cmin, cmax = np.where(c)[0][[0, -1]]
    zmin, zmax = np.where(z)[0][[0, -1]]

    return rmin, rmax, cmin, cmax, zmin, zmax
```

```python
################# CSV GENERATION ############################################

for idx, f in enumerate(file_list):
    patient = np.load(f)['arr_0']
    out_bbox = bbox2_3D(patient[1,])
    zmin, zmax, xmin, xmax, ymin, ymax = out_bbox
    print("%s (%d/%d), %s, (%d, %d)" % (f.split('/')[-1], idx, len(file_list), str(out_bbox), zmin*100.0/patient.shape[1], zmax*100.0/patient.shape[1]))
    names = ['file', 'zmin', 'zmax', 'xmin', 'xmax', 'ymin', 'ymax']
    #print out_bbox
    if xmin>300 or xmax<300:
        print "Error in x!"
    if ymax<300 or ymin>300:
        print "New error y!"
    if zmin*100.0/patient.shape[1]>30 or zmax*100.0/patient.shape[1]<70:
        print "New error z!"
print 'Fin!'
```
