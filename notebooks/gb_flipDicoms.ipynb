```python
import os
import numpy as np
```

```python
path_input = '/mnt/hd2/preprocessed5/'
path_output = '/mnt/hd2/orientation_fix/'
path_file = '/home/gabriel/flippedDCMS.txt'
subselection_path = '/home/gabriel/missingFiles.txt' #Modify the files that are in a subselection only
```

```python
subselection = set()
if subselection_path:
    with open(subselection_path) as f:
        for line in f.readlines():
            subselection.add(line.strip())
else:
    subselection = set(os.listdir(path_input))
```

```python
toFlip = set()
with open(path_file) as f:
    for line in f.readlines():
        toFlip.add(line.split(' ')[1][:-1])

for f in toFlip:
    print "dsb_%s.npz" % f
```

```python
selected = [p for p in toFlip if p in subselection]
# print list(subselection)[0:5]
# print list(toFlip)[0:5]
print selected
```

```python
already_computed = set(os.listdir(path_output))
for i,f in enumerate(os.listdir(path_input)):
    if i % 10 == 0:
        print '%d/%d' % (i, len(os.listdir(path_input)))
    fname = f.split('_')[1].split('.')[0]
    if f.startswith('luna') or fname not in toFlip or fname not in subselection:
        os.system('cp %s%s %s%s' % (path_input, f, path_output, f))
    elif fname in subselection:
        mat = np.load(os.path.join(path_input, f))['arr_0']
        reversed_intensity = mat[0][::-1, :, :]
        reversed_mask = mat[1][::-1, :, :]
        output = np.stack((reversed_intensity, reversed_mask))
        np.savez_compressed(os.path.join(path_output, f), output)
```

```python
for i,fname in enumerate(selected):
    #fname = f.split('_')[1].split('.')[0]
    f = 'dsb_%s.npz' % fname
    print '%s %d/%d' % (f, i, len(selected))
    mat = np.load(os.path.join(path_input, f))['arr_0']
    reversed_intensity = mat[0][::-1, :, :]
    reversed_mask = mat[1][::-1, :, :]
    output = np.stack((reversed_intensity, reversed_mask))
    np.savez_compressed(os.path.join(path_output, f), output)
```
